```python
# Two Sum (1.)

def twoSum(nums, target):
    """
    :type nums: List[int]
    :type target: int
    :rtype: List[int]
    """
    mem = {}
    for i in range(len(nums)):
        if target - nums[i] in mem:
            return [mem[target - nums[i]], i]
        if nums[i] not in mem:
            mem[nums[i]] = i
    
print(twoSum([1,2,3,4,5], 6))
```

```python
import numpy as np

class MyModel:

    def __init__(self, _in, _out):
        self.weights = np.ones([_in,_out])
    
    def forward(self, x):
        return np.matmul(x, self.weights)

model = MyModel(5,5)

print(model.forward(np.array([1,2,3,4,5])))
```
